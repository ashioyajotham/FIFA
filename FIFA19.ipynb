```python
# See the top 20 nationalities in terms of the number of players. 
# value_counts function automatically sort the values.
nationality_count = fifa['Nationality'].value_counts()
nationality_count[:20]
```

```python
fifa.drop("Unnamed: 0", axis=1)
```

```python
# Distribution of age of top nationalities, and clubs based on the number of players.
top_20_nat= nationality_count.index[:20]
df_nationality = fifa[fifa.Nationality.isin(top_20_nat)]

clubs_count = fifa['Club'].value_counts()
top_20_club = clubs_count[:20].index
df_clubs = fifa[fifa.Club.isin(top_20_club)]
```

```python
pd.DataFrame({'Nationality': df_nationality['Age'], 
              'Club': df_clubs['Age']}).hist(bins = 15,
                                                  figsize=(12, 6),
                                                  grid=False,
                                                  rwidth=0.9,
                                                  sharex=True,
                                                  sharey=True)
plt.show()
```

```python
df_nationality['Value'] = df_nationality['Value'].astype(str)
df_nationality['Wage'] = df_nationality['Wage'].astype(str)
```

```python
# Define function to extract the number value from the string
def func(s):
        return eval(s.replace("€", " ").replace('M', "*1000000").replace('K', '*1000').replace("B","*1000000000"))
```

```python
df_nationality['Value'] = df_nationality['Value'].apply(func)
df_nationality['Wage'] = df_nationality['Wage'].apply(func)
```

```python
df_nationality['Value'] = df_nationality['Value'].astype(float)
df_nationality['Wage'] = df_nationality['Wage'].astype(float)
```

```python
df_nationality = df_nationality.rename(columns={"Wage":"Wage(€)", "Value":"Value(€)"})
```

```python
# Estimation of wages for each nationality using the top 20 nationalities DataFrame we created
ax = sns.catplot(x = 'Nationality', y = 'Wage(€)', data = df_nationality,
                hue = 'Preferred Foot', height=6, aspect=2,
                capsize=0.2, kind='point')
plt.xlabel("Nationality", fontsize=12)
plt.ylabel("Wage (EUR)", fontsize=12)
plt.title("Point Estimation of Wages per Top 20 Nationalities", fontsize=20)
plt.xticks(rotation = 60, fontsize=13)
plt.show()
```
